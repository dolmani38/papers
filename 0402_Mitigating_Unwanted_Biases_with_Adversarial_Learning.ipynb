+ <b>Proposition 5.2.</b> Let the training data be comprised of triples(X,Y,Z) drawn according to some distribution D. Suppose:

> (1) The protected variable Z is discrete. -> 예: sex, 남성/ 여성

> (2) The adversary is trained for demographic parity; i.e. the adversary is given only the prediction yˆ. -> adversary는 인구통계학적 동등성만 학습한다. 즉, adversary에는 y^ 만 입력된다(?)

> (3) The adversary is strong enough that, at convergence, it has learned a randomized function A that minimizes the cross entropy loss $ E_{(x,y,z)∼D}[− log P(A(\hat{y}) = z)] $; i.e. the adversary in fact achieves the optimal accuracy with which you can predict Z from Yˆ -> adversary의 function A 는 위의 cross entropy loss 를 갖으며, 충분히 convergence 하다.

> (4) The predictor completely fools the adversary; in particular, the adversary achieves loss H(Z), the entropy of Z. -> predictor f는 adversary A를 완전히 속인다. 

+ Then the predictor satisfies demographic parity; i.e., Yˆ ⊥ Z.
+ 위의 논거들을 만족하면 predictor f는 인구통계학적 동등성을 만족한다. 즉, y^가 z와 independent (성별에 영향을 받지 않는다.)

+ <b>Proposition 5.3.</b> If assumptions (2)-(4) above are replaced with the analogous equality of odds assumptions; in particular, that the adversary is given yˆ and y, and the adversary cannot achieve loss better than H(Z |Y) then the predictor will satisfy Eqality of Odds; i.e., (Yˆ ⊥ Z)|Y
+ 논거 5.2의 (2)~(4)에서 '인구통계학적 동등성'을 '다름에 대한 평등 (equality of odds를 어떻게 해석할지 모르겠음.)' 으로 대체해도 adversary의 loss는 H(Z |Y) 보다 나은것이 없다. 즉, 주어진 y의 조건에서 y^가 z와 independent (성별에 영향을 받지 않는다.)

+ The discriminator for fairness in FR-Train distinguishes predictions w.r.t. one sensitive group from those in the others. 
+ Disparate impact intends the sensitive attribute to be independent of the model’s prediction, i.e., $I(Z;\hat{Y}) = 0$.

+ 여기서, $I(Z;\hat{Y}) = 0$ sensitive group Z가 Y^와 독립적이여햐 한다는 것이고, 내가 의도하는 importance (=sensitive)의 변화는 train dataset을 단위 기간 순서로 자른 N개의 Bucket 또는 어떤 시간 단위 순서의 조건에 대해 독립이여야 한다는 것.
+ 따라서,  N개의 bucket을 B라고 한다면, conditional mutual importance $I(Z;\hat{Y}|B) = 0$ 가 되어야 한다.

## Approach #1


+ 여기서 $ b \in \{1,2,3... N \} = B $
+ https://en.wikipedia.org/wiki/Conditional_mutual_information 를 참조하면 
+ the chain rule for mutual information 에 의해서 $I(Z;\hat{Y}|B)$ 는 다음과 같이 나타낼 수 있다.
$$
I(Z;\hat{Y}|B) = I(Z;\hat{Y}) - I(Z;B) + I(Z;B|\hat{Y})
\tag{1}
$$
+ 여기서, $ \hat{Y}$ 가 $ \{0,1\} \in \hat{Y} $ 이면, 
$$
I(Z;\hat{Y}|B) = I(Z;\hat{Y}) - I(Z;B) + p(\hat{y}=0) I_{\hat{y}=0}(Z_{z \sim \hat{y}=0};B_{b \sim \hat{y}=0}) + p(\hat{y}=1)I_{\hat{y}=1}(Z_{z \sim \hat{y}=1};B_{b \sim \hat{y}=1})
\tag{2}
$$

$$
I(Z;\hat{Y}|B) = I(Z;\hat{Y}) - I(Z;B) + \sum_{\hat{y} \in \hat{Y}}^{0,1} p(\hat{y}) I_{\hat{y}}(Z_{z \sim \hat{y}};B_{b \sim \hat{y}}) 
\tag{3}
$$

+ 여기서, 식(3)의 3번째 항은, conditional probobility (https://en.wikipedia.org/wiki/Conditional_probability) 의 Partial conditional probability 및 "Estimating the Mutual Information between Two Discrete, Asymmetric Variables with Limited Samples" 의 식(2) 에서 추정 함.
+ The conditional entropy H(Y|X) is the average uncertainty of the variable Y once X is known,

I(Z;B) = \sum_{\hat{y} \in \hat{Y}}^{0,1} p(\hat{y}) I_{\hat{y}}(Z_{z \sim \hat{y}};B_{b \sim \hat{y}})

## Approach #2

+ Bucket을 N등분 했을 경우, $ b \in \{1,2,3... N \} = B 이고 $ $ I(Z;\hat{Y}|B) $ 은 각 bucket에서의 $ I_{b}(Z;\hat{Y}) $ 의 조합이다.

$$
I(Z;\hat{Y}|B) = \sum_{b \in B}^N p(b) I(Z_{z \sim b};\hat{Y}_{\hat{y} \sim b})
\tag{4} 
$$
+ p(b) = 1/N 이므로
$$
I(Z;\hat{Y}|B) = \frac{1}{N} \sum_{b \in B}^N I(Z_{z \sim b};\hat{Y}_{\hat{y} \sim b})
\tag{5} 
$$

+ 식(5) 가 0 이되기 위해서, 그리고 RF-Train의 fairness를 만족하기 위해서는 결국 $I(Z_{z \sim b};\hat{Y}_{\hat{y} \sim b}) = 0 $ 이여야 할 것이고, 이는 Approach #1과 동일하게 Theorem1 과 같아 진다? 

+ 한마디로 말해서, RF-Train의 fairness의 $ I(Z;\hat{Y}) = 0 $ 조건은 $ I(Z;\hat{Y}|B) $ 의 조건을 포함 한다.

## RF-Train 에서 도출한 loss function

 + For the loss function of the generator, we employ cross entropy:

$$
\begin{align}
L_1= \dfrac{1}{m} \sum_{i=1}^{m} -y^{(i)} log \hat{y}^{(i)} - (1 - y^{(i)})log(1-\hat{y}^{(i)}) 
\end{align}
$$

+ We set the loss function w.r.t. the fairness discriminator as:

$$
\begin{align}
L_2=  \max_{D(.)} \sum_{z \in Z} \sum_{i:z^{(i)}=z}  \dfrac{1}{m} log D_z(\hat{y}^{(i)}) + H(Z) 
\end{align}
$$

+ The final objective function is the weighted sum of these
value functions:

$$
\begin{align}
\min_{G(.)} L_1 + λ_1 L_2
\end{align}
$$

# 실험1

+ 위의 "RF-Train 에서 도출한 loss function" 에 준하여 실험한다.

L_2 = MSE(\hat{y},Z)
$$

$$
\begin{align}
\min_{G(.)} L_1 - λ_1 L_2
\end{align}
$$

+ 위에서 $λ_1$ 앞이 + 에서 -로 바뀐 것은, MSE를 크게하기 위해서 전체 loss 함수에서는 -로 바꾼 것.
